```python
import time
import pandas as pd
from matplotlib import pyplot as plt

start_time = time.time()

data = pd.read_csv("loans.csv")

columns_list = data.columns.tolist()

print("---Названия столбцов---")
print(columns_list)

nulls_list = []

for i in columns_list:
    nulls_list.append(data[str(i)].isnull().sum())

print("'\n---Общее количество пропусков по столбцам---")
print(nulls_list)

# end_time = time.time()

# print(f"Time of work: {end_time - start_time}")

print("\n---Доля пропусков по столбцам---")

null_data = []
for i, j in zip(columns_list, nulls_list):
    print(f"{i} - {(j/data[str(i)].count()):.4f}")
    null_data.append(j/data[str(i)].count())


print(null_data)

df = pd.DataFrame([null_data], columns=columns_list, index=[1])
df.head()

plt.imshow(df, cmap="RdYlBu")
plt.colorbar()
plt.yticks(range(len(columns_list)), columns_list)
plt.xticks(range(len(null_data)), null_data)
plt.show()
```
