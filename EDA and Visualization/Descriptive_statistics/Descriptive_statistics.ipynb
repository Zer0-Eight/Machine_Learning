```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=0.0, scale=1.0, size=1000)

plt.hist(data, bins=50, density=True, alpha=0.8, color='skyblue', edgecolor='black', label='Гистограмма данных')

x = np.linspace(data.min(), data.max(), 100)
y = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x**2) 

plt.plot(x, y, color='red', linewidth=2, label='Теоретическая плотность')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.title('Гистограмма и плотность нормального распределения')
plt.grid(True, linestyle='--', alpha=0.6)

var_mean = np.sum(data)/len(data)
plt.axvline(var_mean, color='orange', linewidth=1, label='Среднее')
print(var_mean)

var_median = np.median(data)
print(var_median)

plt.axvline(var_mean, color='green', linewidth=1, label='Медиана')

plt.legend()
plt.show()
```
